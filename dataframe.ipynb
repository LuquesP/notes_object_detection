```python
def lables_to_id(df): 
    class_names = df['class_name'].unique() 
    class_names_id = [] 
    for i, c in enumerate(class_names): 
        class_names_id.append({'Id': i, 'Name': c})
    return class_names_id
    
```

```python
def get_label_id(label_dict, label): 
    for e in label_dict: 
        if e['Name'] == label:
            return e['Id'] 
        
```

```python
label_dict = lables_to_id(df) 
print(label_dict[10]['Name'])
```
